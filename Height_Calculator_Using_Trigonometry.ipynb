Trigonometry Height Calculator
Instructions:
1. Press the RUN button
2. There will be a small box where you can enter numbers
3. Enter the values and press enter key
4. It will automatically calculate the values and create a visualization

```python
import math
import matplotlib.pyplot as plt
import numpy as np

elevation_angle = float(input("Enter Angle of Elevation: "))
eye_height = float(input("Enter eye level: "))
object_dist = float(input("Enter distance from object: "))

object_height = (math.tan(math.radians(elevation_angle)) * object_dist) + eye_height
print("Object is " + str(round(object_height, 3)) + " meters tall")

# Generating Graph
cm = 1/2.54  # centimeters in inches
plt.subplots(figsize=(math.ceil(object_dist)*cm, math.ceil(object_height)*cm))
plt.xlim(0, math.ceil(object_dist))
plt.ylim(0, math.ceil(object_height))
plt.grid(linewidth=1.25)
plt.grid(which='minor', linewidth=0.25)
plt.minorticks_on()
vertices = [(0, eye_height), (object_dist, round(object_height, 3)), (object_dist, eye_height), (0, eye_height)]
rectvertices = [(0, eye_height), (0, 0), (object_dist, 0), (object_dist, eye_height)]
x, y = zip(*vertices)
l, b = zip(*rectvertices)
plt.plot(x, y, marker='o', linestyle='-', color='orange')
plt.plot(l, b, marker='o', linestyle='-', color='red')
plt.text(0.5 + object_dist, math.ceil(object_height)/2, "BC = " + str(round(object_height, 3) - eye_height) + "m")
plt.text(0.5 + object_dist, math.ceil(object_height)/2 - 1, "    BE(Object height) = " + str(round(object_height, 3)) + "m")
plt.text(0.5 + object_dist, (eye_height)/2 - 0.25, "CE = " + str(eye_height) + "m")
plt.text((object_dist/2) - 0.5, 0.25, "DE = " + str(object_dist))

plt.text(0.25, eye_height - 0.5, "A")
plt.text(object_dist + 0.25, math.ceil(object_height) - 0.5, "B")
plt.text(object_dist - 0.5, eye_height - 0.5, "C")
plt.text(0.25, 0.25, "D")
plt.text(object_dist - 0.5, 0.25, "E")

plt.text(object_dist - 1.25, math.ceil(object_height) - 2, s=str(180 - (elevation_angle + 90)) + "°")
plt.text(object_dist - 1, eye_height + 0.25, "90°")
plt.text(1, eye_height + 0.25, s=str(elevation_angle) + "°")


# Add labels and title
plt.xlabel('Distance from Object')
plt.ylabel('Height of Object')
plt.title('Trigonometry Height Calculator')

# Show the plot
plt.axis('equal')
plt.show()
```
